### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
import math
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

```python
def f(x):
    return x**5 + 4*math.sin(2*x) + math.cos(3*x + 3)

print(derivative(f, 1.0, dx=1e-6, n=1))
print(derivative(f, 1.0, dx=1e-6, n=2))
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def f(x):
    return (math.sin(2*x + 1))**5

x = -5
d1 = []
d2 = []
indices = []
while x < 5:
    d1.append(derivative(f, x, dx=1e-6, n=1))
    d2.append(derivative(f, x, dx=1e-6, n=2))
    indices.append(x)
    x = x + 1e-3

plt.plot(indices, d1, label='d1')
plt.plot(indices, d2, label='d2')
plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
